```python
import pandas as pd

df=pd.read_csv("event_filtered.csv")
```

```python
df.columns
```

```python
df["matchname"].unique()
```

```python
import datetime as dt
df["matchdate"]=pd.to_datetime(df["matchdate"])
```

```python

df[['hometeam', 'awayteam']] = df['matchname'].str.split('vs', expand=True)
df['hometeam'] = df['hometeam'].str.strip()
df['awayteam'] = df['awayteam'].str.strip()
df["half"] = df["timemin"].apply(lambda x: "first" if x <= 45 else "second")
```

```python
def categorize_shot(row):
    if row["goal"] == 1:
        return "goal"
    elif row["shotontarget"] == 1:
        return "on target"
    else:
        return "off target"

df["shot_result"] = df.apply(categorize_shot, axis=1)
```

```python
colors = {"goal": "red", "on target": "blue", "off target": "black"}
```

```python
import matplotlib.pyplot as plt
from mplsoccer import VerticalPitch, Pitch
```

```python
team_cn=sorted(df["contestantname"].unique())
teams=sorted(df["hometeam"].unique())
print(team_cn)
print(teams)
team_mapping = dict(zip(team_cn, teams))
print(team_mapping)
```

```python
df["contestantname"] = df["contestantname"].map(team_mapping)
```

```python
df[["matchname", "contestantname", "hometeam", "awayteam"]].head(8)
```

```python
import pandas as pd
from mplsoccer import VerticalPitch
import matplotlib.pyplot as plt
from ipywidgets import interact, Dropdown

pitch = VerticalPitch(pitch_type='uefa',  half=True, line_color='black')

def plot_shots(team, home_away, scored_conceded, shot_type, half):
    fig, ax = pitch.draw(figsize=(10, 20))
    applied_filters = []
    filtered = df.copy()

    # --- team FILTER ---
    filtered = filtered[(filtered["hometeam"] == team) | (filtered["awayteam"] == team)]
    applied_filters.append(team)

    #home/away
    if home_away == "Home":
        filtered = filtered[filtered["hometeam"] == team]
    elif home_away == "Away":
        filtered = filtered[filtered["awayteam"] == team]
    applied_filters.append(home_away)

    #scored/conceded
    if scored_conceded == "Scored":
        filtered = filtered[filtered["contestantname"] == team]
    elif scored_conceded == "Conceded":
        filtered = filtered[filtered["contestantname"] != team]
    applied_filters.append(scored_conceded)

    # --- SHOT TYPE FILTER ---
    if shot_type == "On target":
        filtered = filtered[filtered["shot_result"] == "on target"]
    elif shot_type == "Off target":
        filtered = filtered[filtered["shot_result"] == "off target"]
    elif shot_type == "Goal":
        filtered = filtered[filtered["shot_result"] == "goal"]
    applied_filters.append(shot_type)

    # --- HALF FILTER ---
    if half == "1st half":
        filtered = filtered[filtered["timemin"] <= 45]
    elif half == "2nd half":
        filtered = filtered[filtered["timemin"] > 45]
    applied_filters.append(half)
    # --- CALCULATIONS ---
    count = len(filtered)
    if count > 0:
        x_avg = filtered["x"].mean()
        y_avg = filtered["y"].mean()
        xg_avg = filtered["expectedgoals"].mean()
    else:
        avg_text = "-"
    # --- DRAW SHOTS ---
    for outcome, color in colors.items():
        subset = filtered[filtered["shot_result"] == outcome]
        if subset.empty:
            continue
        if outcome == "off target":
            pitch.scatter(
                subset.x, subset.y,
                s=subset.expectedgoals * 1000,
                facecolors="none",
                edgecolors=color,
                linewidth=2,
                ax=ax,
            )
        else:
            pitch.scatter(
                subset.x, subset.y,
                s=subset.expectedgoals * 1000,
                edgecolor="grey",
                color=color,
                ax=ax,
            )
    pitch.scatter(x_avg, y_avg, edgecolor="green",color="#49F527", s=100, ax=ax)
    plt.title(f"SHOTMAP\n2023/24 OTP Bank Liga NB1 - {team}", fontsize=16)
    # label
    ax.text(
        0, 1.1,
        f"Filtered shots: {count}\nFilters: {', '.join(applied_filters)}\nAvg. xG: {xg_avg.round(3)}",
        ha="left", va="bottom",
        fontsize=12, color="black",
        transform=ax.transAxes,
    )
    ax.text(
        0, 1.07,
        f"Avg. position: x: {x_avg.round(1)}, y: {y_avg.round(1)}",
        ha="left", va="bottom",
        fontsize=12, color="#49F527",
        transform=ax.transAxes,
    )
    plt.show()


# --- WIDGETS ---


interact(
    plot_shots,
    team=Dropdown(options=teams, value=teams[0],
                  description="Team"),
    scored_conceded=Dropdown(options=["Scored", "Conceded"], value="Scored",
                 description="Scored/Conceded:"),
    home_away=Dropdown(options=["All","Home", "Away"], value="All",
                       description="Home/Away:"),
    shot_type=Dropdown(options=["All shots", "On target", "Off target", "Goal"], value="All shots",
                       description="Shot Type:"),
    half=Dropdown(options=["Full time", "1st half", "2nd half"], value="Full time",
                  description="Half:")
)
```
