```python
import pandas as pd
```

```python
df=pd.read_excel("felv_shotmap_input_v00.xlsx")
df
```

```python

paks = df[df['matchname'].str.startswith('Paks')]
#paks
```

```python
def categorize_shot(row):
    if row["goal"] == 1:
        return "goal"
    elif row["shotontarget"] == 1:
        return "on target"
    else:
        return "off target"

paks["shot_result"] = paks.apply(categorize_shot, axis=1)
```

```python
shots=paks[["expectedgoals", "x", "y", "shot_result"]].dropna(subset=["expectedgoals"])
#shots
```

```python
colors = {"goal": "red", "on target": "blue", "off target": "black"}
```

```python
import matplotlib.pyplot as plt
from mplsoccer import VerticalPitch, Pitch
```

```python
pitch = VerticalPitch(line_color='black', half = True)
fig, ax = pitch.grid(grid_height=0.9, title_height=0.06, axis=False,
                     endnote_height=0.04, title_space=0, endnote_space=0)

for outcome, color in colors.items():
    subset = shots[shots["shot_result"] == outcome]
    if not subset.empty:
        if outcome == "off target":
            pitch.scatter(subset.x, subset.y,
                          s=subset.expectedgoals * 1000,
                          facecolors="none",
                          edgecolors=color,
                          linewidth=2,
                          ax=ax['pitch'])
        else:
            pitch.scatter(subset.x, subset.y,
                          s=subset.expectedgoals * 1000,
                          color=color,
                          edgecolors="black",
                          ax=ax['pitch'])
fig.suptitle("Paks lövései", fontsize=20, fontweight="bold")
ax['title'].text(0.5, -0.4, "NB I 2023-2024", fontsize=12, ha="center", va="top")
plt.show()
```
